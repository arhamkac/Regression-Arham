```python
import pandas as pd
import numpy as np
```

```python
wines=pd.read_csv("wines_SPA.csv")
wines.head()
```

```python
wines.isna().sum()
```

```python
wines.drop('year',axis='columns',inplace=True)
```

```python
wines['body'].fillna(wines['body'].median(),inplace=True)
wines['acidity'].fillna(wines['acidity'].median(),inplace=True)
```

```python
wines.drop('type',axis='columns',inplace=True)
```

```python
Q1 = wines['num_reviews'].quantile(0.25)
Q3 = wines['num_reviews'].quantile(0.75)
IQR = Q3 - Q1
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR
wines_f = wines[(wines['num_reviews'] > lower_bound) & (wines['num_reviews'] < upper_bound)]
```

```python
#Converting datatype to all num
from sklearn.preprocessing import LabelEncoder
le=LabelEncoder()
wines_f['winery']=le.fit_transform(wines_f['winery'])
wines_f['wine']=le.fit_transform(wines_f['wine'])
wines_f['country']=le.fit_transform(wines_f['country'])
wines_f['region']=le.fit_transform(wines_f['region'])
```

```python
wines_f.head()
```

```python
#Train Test Split Functionality
x=wines_f.drop('price',axis='columns')
y=wines_f['price']
def train_test_split(x,y,test_size=0.2):
    #Shuffled using random.shuffle method in numpy by first changing the indexing
    #of elements and then locating them with their indexes
    index=np.arange(len(x))
    np.random.shuffle(index)
    x=x.iloc[index]
    y=y.iloc[index]
    split_size=int(len(x)*(1-test_size))
    x_train=x[:split_size]
    x_test=x[split_size:]
    y_train=y[:split_size]
    y_test=y[split_size:]
    return x_train,x_test,y_train,y_test
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.25)
```
